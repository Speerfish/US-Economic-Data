# Step 1: Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load the CSV file into a pandas DataFrame

```python
df = pd.read_csv('PCU5311205311201.csv', parse_dates=['DATE'])
df['PCU5311205311201'] = pd.to_numeric(df['PCU5311205311201'], errors='coerce')
```

## Step 3: Inspect the data

```python
print(df.info())
df.head(15)  # Display the first few rows
```

## Step 4: Plot time series line graph

```python
# Plot the time series data
plt.figure(figsize=(14, 8))
sns.lineplot(x='DATE', y='PCU5311205311201', data=df)

# Add labels and title
plt.title(' Producer Price Index by Industry: Lessors of Nonresidential Buildings: Leasing of Shopping Centers and Retail Stores Over Time')
plt.xlabel('Date')
plt.ylabel('Index Dec 1995=100')

# Display the plot
plt.grid(True)
plt.show()
```
