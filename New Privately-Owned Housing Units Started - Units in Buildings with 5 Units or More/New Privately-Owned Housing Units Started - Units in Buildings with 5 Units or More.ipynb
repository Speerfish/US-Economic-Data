# Step 1: Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load the CSV file into a pandas DataFrame

```python
df = pd.read_csv('HOUST5F.csv', parse_dates=['DATE'])
df['HOUST5F'] = pd.to_numeric(df['HOUST5F'], errors='coerce')
```

## Step 3: Inspect the data

```python
print(df.info())
df.head(15)  # Display the first few rows
```

## Step 4: Plot time series line graph

```python
# Plot the time series data
plt.figure(figsize=(12, 8))
sns.lineplot(x='DATE', y='HOUST5F', data=df)

# Add labels and title
plt.title('New Privately-Owned Housing Units Started: Units in Buildings with 5 Units or More Over Time')
plt.xlabel('Date')
plt.ylabel('Thousands of Units')

# Display the plot
plt.grid(True)
plt.show()
```
